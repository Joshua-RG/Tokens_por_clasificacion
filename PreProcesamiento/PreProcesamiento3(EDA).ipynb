```python
def plot_marketcap_trend(halving_data_list, halving_labels, halving_dates):
    # Crear una lista para almacenar las medianas de Market Cap
    marketcap_medians = []

    # Calcular la mediana de Market Cap para cada dataset
    for data in halving_data_list:
        median_value = data['Market_Cap'].median()
        marketcap_medians.append(median_value)

    # Crear un DataFrame para las fechas y medianas
    marketcap_df = pd.DataFrame({
        'Fecha': halving_dates,
        'Mediana_Market_Cap': marketcap_medians
    })

    # Crear gráfico de líneas
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=marketcap_df, x='Fecha', y='Mediana_Market_Cap', marker='o')
    plt.title('Tendencia de la Mediana de la Capitalización de Mercado a lo largo de los Halvings')
    plt.xlabel('Fecha')
    plt.ylabel('Mediana de la Capitalización de Mercado (USD)')
    plt.yscale('log')  # Escala logarítmica para mejor visualización
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.show()

# Lista de fechas asociadas a cada CSV (puedes ajustar estas fechas según corresponda)
halving_dates = ['2012-11-28', '2013-08-06',  # Halving #1 y #1 + 250 Días
                 '2016-07-09', '2017-03-16',  # Halving #2 y #2 + 250 Días
                 '2020-05-11', '2021-01-16',  # Halving #3 y #3 + 250 Días
                 '2024-01-01']  # Halving #4

# Llamar a la función con los datasets
halving_data_list = [halving_1, halving_1_plus250, halving_2, halving_2_plus250, halving_3, halving_3_plus250, halving_4]
halving_labels = ['Halving #1', 'Halving #1 + 250 Días', 'Halving #2', 'Halving #2 + 250 Días', 'Halving #3', 'Halving #3 + 250 Días', 'Halving #4']

plot_marketcap_trend(halving_data_list, halving_labels, halving_dates)
```

```python
def plot_marketcap_comparison(halving_data_list, halving_labels):
    # Crear una lista para almacenar las medianas de Market Cap
    marketcap_medians = []

    # Calcular la mediana de Market Cap para cada dataset
    for data in halving_data_list:
        median_value = data['Market_Cap'].median()
        marketcap_medians.append(median_value)

    # Crear gráfico de barras
    plt.figure(figsize=(10, 6))
    sns.barplot(x=halving_labels, y=marketcap_medians)
    plt.title('Comparación de la Mediana de la Capitalización de Mercado entre Halvings')
    plt.xlabel('Halving')
    plt.ylabel('Mediana de la Capitalización de Mercado (USD)')
    plt.yscale('log')  # Escala logarítmica para mejorar la visualización
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

# Llamar a la función con los datasets
halving_data_list = [halving_1, halving_1_plus250, halving_2, halving_2_plus250, halving_3, halving_3_plus250, halving_4]
halving_labels = ['Halving #1', 'Halving #1 + 250 Días', 'Halving #2', 'Halving #2 + 250 Días', 'Halving #3', 'Halving #3 + 250 Días', 'Halving #4']

plot_marketcap_comparison(halving_data_list, halving_labels)
```

```python
def plot_volume_vs_market_cap(halving_data, halving_label):
    plt.figure(figsize=(12, 6))
    sns.scatterplot(data=halving_data, x='Volume_24h', y='Market_Cap', alpha=0.7)
    plt.title(f'Relación entre Volumen y Capitalización de Mercado - {halving_label}')
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel('Volumen en 24h (USD)')
    plt.ylabel('Capitalización de Mercado (USD)')
    plt.show()

# Graficar la relación entre volumen y capitalización de mercado para cada dataset de halving
plot_volume_vs_market_cap(halving_1, 'Halving #1')
plot_volume_vs_market_cap(halving_1_plus250, 'Halving #1 + 250 Días')
plot_volume_vs_market_cap(halving_2, 'Halving #2')
plot_volume_vs_market_cap(halving_2_plus250, 'Halving #2 + 250 Días')
plot_volume_vs_market_cap(halving_3, 'Halving #3')
plot_volume_vs_market_cap(halving_3_plus250, 'Halving #3 + 250 Días')
plot_volume_vs_market_cap(halving_4, 'Halving #4')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset principal
main_data = pd.read_csv('InformationByTokenClass.csv')

# Crear una nueva columna para el Halving correspondiente
def assign_halving(row):
    if row['Halving_1_RankIndex'] > 0:
        return 'Halving #1'
    elif row['Halving_1_Plus250_RankIndex'] > 0:
        return 'Halving #1 + 250 días'
    elif row['Halving_2_RankIndex'] > 0:
        return 'Halving #2'
    elif row['Halving_2_Plus250_RankIndex'] > 0:
        return 'Halving #2 + 250 días'
    elif row['Halving_3_RankIndex'] > 0:
        return 'Halving #3'
    elif row['Halving_3_Plus250_RankIndex'] > 0:
        return 'Halving #3 + 250 días'
    else:
        return 'No Clasificado'

# Aplicar la función para crear la nueva columna
main_data['Halving'] = main_data.apply(assign_halving, axis=1)

# Filtrar solo los datos clasificados
halving_data = main_data[main_data['Halving'] != 'No Clasificado']

# Establecer el orden de los halvings manualmente
halving_order = ['Halving #1', 'Halving #1 + 250 días', 'Halving #2', 'Halving #2 + 250 días', 'Halving #3', 'Halving #3 + 250 días']

# Crear un gráfico de caja para visualizar la distribución de Capitalización de Mercado
plt.figure(figsize=(14, 7))
sns.boxplot(data=halving_data, x='Halving', y='Market_Cap', order=halving_order, palette='Set2')
plt.title('Distribución de Capitalización de Mercado por Halving')
plt.xlabel('Halving')
plt.ylabel('Capitalización de Mercado')
plt.yscale('log')  # Usar escala logarítmica si hay grandes diferencias en los valores
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset principal
main_data = pd.read_csv('InformationByTokenClass.csv')

# Crear una nueva columna para el Halving correspondiente (incluyendo el Halving #4)
def assign_halving(row):
    if row['Halving_1_RankIndex'] > 0:
        return 'Halving #1'
    elif row['Halving_1_Plus250_RankIndex'] > 0:
        return 'Halving #1 + 250 días'
    elif row['Halving_2_RankIndex'] > 0:
        return 'Halving #2'
    elif row['Halving_2_Plus250_RankIndex'] > 0:
        return 'Halving #2 + 250 días'
    elif row['Halving_3_RankIndex'] > 0:
        return 'Halving #3'
    elif row['Halving_3_Plus250_RankIndex'] > 0:
        return 'Halving #3 + 250 días'
    elif row['Halving_4_RankIndex'] > 0:
        return 'Halving #4'
    else:
        return 'No Clasificado'

# Aplicar la función para crear la nueva columna
main_data['Halving'] = main_data.apply(assign_halving, axis=1)

# Filtrar solo los datos clasificados
halving_data = main_data[main_data['Halving'] != 'No Clasificado']

# Realizar el conteo de las criptomonedas por Halving y asegurarse que los halvings están en orden específico
halving_order = ['Halving #1', 'Halving #1 + 250 días', 'Halving #2', 'Halving #2 + 250 días',
                 'Halving #3', 'Halving #3 + 250 días', 'Halving #4']
coin_counts = halving_data['Halving'].value_counts().reindex(halving_order).fillna(0).reset_index()
coin_counts.columns = ['Halving', 'Count']

# Imprimir el conteo de monedas por Halving
print(coin_counts)

# Crear un gráfico de barras para visualizar el conteo de monedas por Halving
plt.figure(figsize=(14, 7))
sns.barplot(data=coin_counts, x='Halving', y='Count', palette='Set2')

# Agregar valores sobre las barras
for index, value in enumerate(coin_counts['Count']):
    plt.text(index, value + 5, str(int(value)), ha='center')

plt.title('Conteo de Criptomonedas por Halving (Del Halving #1 al Halving #4)')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datasets de halving
halving_1 = pd.read_csv('Halving_csv/halving_1.csv')
halving_1_plus250 = pd.read_csv('Halving_csv/halving_1_plus250.csv')
halving_2 = pd.read_csv('Halving_csv/halving_2.csv')
halving_2_plus250 = pd.read_csv('Halving_csv/halving_2_plus250.csv')
halving_3 = pd.read_csv('Halving_csv/halving_3.csv')
halving_3_plus250 = pd.read_csv('Halving_csv/halving_3_plus250.csv')
halving_4 = pd.read_csv('Halving_csv/halving_4.csv')

# Contar el número de criptomonedas en cada halving
halving_counts = {
    'Halving #1': halving_1.shape[0],
    'Halving #1 + 250 días': halving_1_plus250.shape[0],
    'Halving #2': halving_2.shape[0],
    'Halving #2 + 250 días': halving_2_plus250.shape[0],
    'Halving #3': halving_3.shape[0],
    'Halving #3 + 250 días': halving_3_plus250.shape[0],
    'Halving #4': halving_4.shape[0]
}

# Convertir el conteo a un DataFrame
halving_counts_df = pd.DataFrame(list(halving_counts.items()), columns=['Halving', 'Count'])

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(data=halving_counts_df, x='Halving', y='Count', palette='Set2')

# Agregar valores sobre las barras
for index, value in enumerate(halving_counts_df['Count']):
    plt.text(index, value + 1, str(value), ha='center')

plt.title('Conteo de Criptomonedas por Halving')
plt.xlabel('Halving')
plt.ylabel('Número de Criptomonedas')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Mostrar el gráfico
plt.show()
```
