#Importar dependencias necesarias

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

#Carga de datos del CSV principal

```python
# Configuración de visualizaciones
sns.set(style="whitegrid")
plt.style.use('fivethirtyeight')

# Cargar el dataset
df = pd.read_csv('BasicInformationByTokenClass.csv')

# Mostrar las primeras filas del dataset
display(df.head())

# Visualizar los tipos de datos
print("Tipos de datos en el dataset:")
print(df.dtypes)

# Obtener información general del dataset
df.info()
```

#Estadisticas Descriptivas

```python
# Mostrar estadísticas descriptivas para las columnas numéricas
display(df.describe())

# Mostrar estadísticas para las columnas categóricas
display(df.describe(include=['O']))  # 'O' es para objetos, generalmente cadenas de texto
```

#Limpieza de datos

```python
# Eliminar duplicados
df.drop_duplicates(inplace=True)

```

```python
# Renombrar las columnas con mayúsculas y guiones bajos
df = df.rename(columns={
    'Ranking': 'Ranking',
    'Name': 'Name',
    'Token': 'Token',
    'Price': 'Price',
    '1h%': 'One_Hour_Percentage',
    '24h%': 'TwentyFour_Hour_Percentage',
    '7d%': 'Seven_Days_Percentage',
    'Market Cap': 'Market_Cap',
    'Volume(24h)': 'Volume_24h',
    'Circulating Supply': 'Circulating_Supply',
    'Class': 'Class',
    'halving_1(RankIndex)': 'Halving_1_RankIndex',
    'halving_1_plus250(RankIndex)': 'Halving_1_Plus250_RankIndex',
    'halving_2(RankIndex)': 'Halving_2_RankIndex',
    'halving_2_plus250(RankIndex)': 'Halving_2_Plus250_RankIndex',
    'halving_3(RankIndex)': 'Halving_3_RankIndex',
    'halving_3_plus250(RankIndex)': 'Halving_3_Plus250_RankIndex',
    'halving_4(RankIndex)': 'Halving_4_RankIndex',
    'total_supply': 'Total_Supply',
    'max_supply': 'Max_Supply'
})


# Reorganizar las columnas para que las relacionadas con Supply estén cercanas
column_order = [
    'Ranking', 'Name', 'Token', 'Price', 'One_Hour_Percentage',
    'TwentyFour_Hour_Percentage', 'Seven_Days_Percentage', 'Market_Cap','URL','socialNetworks','socialNetworksCount',
    'Volume_24h', 'Circulating_Supply', 'Total_Supply', 'Max_Supply', 'Class',
    'Halving_1_RankIndex', 'Halving_1_Plus250_RankIndex', 'Halving_2_RankIndex',
    'Halving_2_Plus250_RankIndex', 'Halving_3_RankIndex', 'Halving_3_Plus250_RankIndex',
    'Halving_4_RankIndex'
]


# Reorganizar las columnas
df = df[column_order]

# Añadir una nueva columna con la fecha de captura
df['Capture_Date'] = '2024-09-23'


# Definir las columnas a verificar
halving_columns = [
    'Halving_1_RankIndex',
    'Halving_1_Plus250_RankIndex',
    'Halving_2_RankIndex',
    'Halving_2_Plus250_RankIndex',
    'Halving_3_RankIndex',
    'Halving_3_Plus250_RankIndex'
]

# Crear la nueva columna 'halving_survive'
df['halving_survive'] = df[halving_columns].notnull().any(axis=1).astype(int)

# Mostrar las primeras filas del DataFrame para verificar los cambios
df.head()
```

#Guardar dataset

```python
# Guardar el dataset preprocesado
df.to_csv('InformationByTokenClass.csv', index=False)

print("Preprocesamiento completado y guardado.")
```

**Leer CSV de los halving +250 dias**

```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt

# Configuración de visualizaciones
sns.set(style="whitegrid")
plt.style.use('fivethirtyeight')

# Leer los archivos CSV desde la carpeta "Clean"
carpeta_clean = 'Clean/'

archivos = [
    'halving_1_clean.csv',
    'halving_1_plus250_clean.csv',
    'halving_2_clean.csv',
    'halving_2_plus250_clean.csv',
    'halving_3_clean.csv',
    'halving_3_plus250_clean.csv',
    'halving_4_clean.csv'
]

# Procesar cada archivo CSV
for archivo in archivos:
    ruta = os.path.join(carpeta_clean, archivo)

    # Cargar el archivo CSV
    df = pd.read_csv(ruta)

    # Mostrar las primeras filas del dataset
    print(f"\nPrimeras filas del archivo: {archivo}")
    display(df.head())

    # Visualizar los tipos de datos
    print(f"Tipos de datos en el dataset {archivo}:")
    print(df.dtypes)

    # Obtener información general del dataset
    print(f"Información del dataset {archivo}:")
    df.info()

    # Eliminar duplicados si existen
    df.drop_duplicates(inplace=True)

    # Verificar si se han eliminado duplicados
    print(f"Duplicados eliminados en {archivo}, total de filas después: {len(df)}")

    # Renombrar las columnas
    df = df.rename(columns={
        'name': 'Name',
        'symbol': 'Symbol',
        'cmcRank': 'Rank',
        'circulatingSupply': 'Circulating_Supply',
        'totalSupply': 'Total_Supply',
        'maxSupply': 'Max_Supply',
        'dateAdded': 'Date_Added',
        'price': 'Price',
        'volume24h': 'Volume_24h',
        'marketCap': 'Market_Cap',
        'percentChange1h': 'Percent_Change_1h',
        'percentChange24h': 'Percent_Change_24h',
        'percentChange7d': 'Percent_Change_7d'
    })

    # Reorganizar las columnas para que las importantes estén al inicio
    column_order = [
        'Rank', 'Name', 'Symbol', 'Price', 'Volume_24h',
        'Market_Cap', 'Circulating_Supply', 'Total_Supply',
        'Max_Supply', 'Percent_Change_1h', 'Percent_Change_24h',
        'Percent_Change_7d', 'Date_Added'
    ]

    # Reorganizar las columnas
    df = df[column_order]

    # Añadir una nueva columna con la fecha de captura
    df['Capture_Date'] = '2024-09-23'

    # Mostrar las primeras filas del DataFrame para verificar los cambios
    print(f"\nDataFrame modificado del archivo: {archivo}")
    display(df.head())


# Crear carpeta para guardar los archivos CSV procesados
carpeta_halving = 'Halving_csv'
os.makedirs(carpeta_halving, exist_ok=True)

# Guardar los archivos CSV procesados en la nueva carpeta
for archivo in archivos:
    nombre_archivo_salida = archivo.replace('_clean.csv', '.csv')
    ruta_salida = os.path.join(carpeta_halving, nombre_archivo_salida)
    df.to_csv(ruta_salida, index=False)
    print(f"Archivo guardado: {ruta_salida}")
```
